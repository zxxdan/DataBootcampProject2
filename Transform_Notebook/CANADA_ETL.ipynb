```python
import os
import pandas as pd
```

```python
can_data_file = os.path.join("..","Extract","Canada","NCDB_1999_to_2014.csv")
```

```python
canada_df = pd.read_csv(can_data_file, low_memory=False)
canada_df.head()
```

```python
columns_to_keep = ["C_YEAR","C_MNTH","C_SEV","C_VEHS","V_TYPE",
               "P_SEX","P_USER","P_AGE","V_YEAR"
              ]
trimmed_df = canada_df[columns_to_keep]
trimmed_df.head()
```

```python
filtered_df = trimmed_df.loc[(trimmed_df["C_YEAR"]>2008) & (trimmed_df["C_YEAR"]<2015),:]
filtered_df
```

```python
c_sev = {
    1: "Fatal",
    2: "Non-Fatal",
    "U": "Undisclosed",
    "X": "Undisclosed"
}

v_type = {
    '01':"Passenger Vehicle", 
    '20':"Construction Equipment", 
    '05':"Cargo Vehicle", 
    '22':"Recreation Vehicles", 
    'NN':"Undisclosed", 
    '06':"Cargo Vehicle", 
    '07':"Trucks", 
    'QQ':"Undisclosed", 
    '11':"Bus", 
    '08':"Construction Equipment", 
    '17':"Bicycle",
    'UU':"Undisclosed", 
    '09':"Bus", 
    '16':"Recreation Vehicles", 
    '10':"Bus", 
    '21':"Fire Engine",
    '14':"Motorcycle",
    '23':"Tram",
    '19':"Agriculture Machine",
    '18':"Purpose-built motorhome",
    'XX':"Undisclosed"
    
}


p_user = {
    "1":"Motor Vehicle Occupant",
    "2":"Motor Vehicle Occupant",
    "3":"Pedestrian",
    "4":"Cyclist",
    "5":"Motorcyclist",
    "U":"Undisclosed"
}

p_sex = {
    "F":"Female",
    "M":"Male",
    "N": "Undisclosed",
    "U": "Undisclosed",
    "X": "Undisclosed"
}

#C_MNTH,C_VEHS,P_AGE,V_YEAR need to update manually

filtered_df["C_SEV"] = filtered_df["C_SEV"].map(c_sev)
filtered_df["V_TYPE"] = filtered_df["V_TYPE"].map(v_type)
filtered_df["P_USER"] = filtered_df["P_USER"].map(p_user)
filtered_df["P_SEX"] = filtered_df["P_SEX"].map(p_sex)


filtered_df.loc[filtered_df['C_MNTH'] == "UU", ['C_MNTH']] = 'Undisclosed'
filtered_df.loc[filtered_df['C_VEHS'] == "UU", ['C_VEHS']] = 'Undisclosed'
filtered_df.loc[filtered_df['C_VEHS'] == "XX", ['C_VEHS']] = 'Undisclosed'
filtered_df.loc[filtered_df['P_AGE'] == "UU", ['P_AGE']] = 'Undisclosed'
filtered_df.loc[filtered_df['P_AGE'] == "NN", ['P_AGE']] = 'Undisclosed'
filtered_df.loc[filtered_df['V_YEAR'] == "NNNN", ['V_YEAR']] = 'Undisclosed'
filtered_df.loc[filtered_df['V_YEAR'] == "UUUU", ['V_YEAR']] = 'Undisclosed'
```

```python
undisclosed_months = filtered_df[filtered_df["C_MNTH"] == "Undisclosed"].index
undisclosed_age = filtered_df[filtered_df["P_AGE"] == "Undisclosed"].index

filtered_df.dropna(inplace=True)
filtered_df.drop(undisclosed_months, inplace = True)
filtered_df.drop(undisclosed_age, inplace = True)
```

```python
filtered_df = filtered_df.rename(columns = {"C_YEAR":"Collision_Year",
                                            "C_MNTH":"Collision_Month",
                                            "C_SEV":"Collision_Severity",
                                            "C_VEHS":"Number_of_Vehicles",
                                            "V_TYPE":"Vehicle_Type",
                                            "P_SEX":"Person_Gender",
                                            "P_USER": "Person_Position",
                                            "P_AGE": "Person_Age",
                                            "V_YEAR": "Vehicle_Year"
})
```

```python
filtered_df.to_csv(os.path.join("..","Staging","Canada_Final.csv"), index = False, header=True)
```
