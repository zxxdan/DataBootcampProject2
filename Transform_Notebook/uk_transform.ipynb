```python
import os
import pandas as pd
import numpy as np
import datetime
```

```python
# Paths to the main UK files

uk_accident_file = os.path.join("..","Extract","UK","Accidents0515.csv")
uk_casualty_file = os.path.join("..","Extract","UK","Casualties0515.csv")
uk_vehicle_file = os.path.join("..","Extract","UK","Vehicles0515.csv")

# Additional context files needed
uk_casu_class = os.path.join("..","Extract","UK","contextCSVs","Casualty_Class.csv") #for column casualty class
uk_sex = os.path.join("..","Extract","UK","contextCSVs","Sex_of_Driver.csv") #for casualty sex, and driver sex
uk_casu_sev = os.path.join("..","Extract","UK","contextCSVs","Casualty_Severity.csv") #for casualty severity
uk_casu_type = os.path.join("..","Extract","UK","contextCSVs","Casualty_Type.csv") #for casualty type
uk_vehi_type = os.path.join("..","Extract","UK","contextCSVs","Vehicle_Type.csv") # for vehicle type
```

```python
# Loading accident file as panda dataframe

uk_accident_df = pd.read_csv(uk_accident_file, low_memory=False)
uk_accident_df.head()
```

```python
# Loading casualty file as panda dataframe

uk_casualty_df = pd.read_csv(uk_casualty_file)
uk_casualty_df.head()
```

```python
# Loading vehicle file as panda dataframe

uk_vehicle_df = pd.read_csv(uk_vehicle_file)
uk_vehicle_df.head()
```

```python
# Merging the accident and casualty dataframes

uk_acci_casu_df = pd.merge(uk_casualty_df,uk_accident_df,on="Accident_Index",how="left")
uk_acci_casu_df.head()
```

```python
# Merging the new accident + casualty dataframe with the vehicle dataframe

uk_acci_casu_vehi_df = pd.merge(uk_acci_casu_df,uk_vehicle_df,on="Accident_Index",how="left")
uk_acci_casu_vehi_df.head()
```

```python
# Determining the columns we want to focus on, based on what is also available in the Canada data

uk_data_df = uk_acci_casu_vehi_df[['Accident_Index',
                                   'Casualty_Class',
                                   'Sex_of_Casualty',
                                   'Age_of_Casualty',
                                   'Casualty_Severity',
                                   'Casualty_Type',
                                   'Number_of_Vehicles',
                                   'Number_of_Casualties',
                                   'Date',
                                   'Vehicle_Type',
                                   'Sex_of_Driver',
                                   'Age_of_Driver',
                                   'Age_of_Vehicle'
                                  ]]

uk_data_df.head()
```

```python
# Changing date format by changing type to datetime

uk_data_df["Date"] = uk_data_df["Date"].astype('datetime64')
```

```python
# Extracting and creating new column for year and month seperately

uk_data_df['Year'] = pd.DatetimeIndex(uk_data_df["Date"]).year
uk_data_df['Month'] = pd.DatetimeIndex(uk_data_df["Date"]).month
uk_data_df.head()
```

```python
#Determining the year of the vehicle in the collision, based on the age of vehicle and year of accident

vehi_year = uk_data_df['Year']-uk_data_df['Age_of_Vehicle']
uk_data_df["Vehicle_Year"] = vehi_year
uk_data_df.head()
```

```python
# Determining the columns we want to focus on, based on what is also available in the Canada data

uk_data_df = uk_data_df[['Accident_Index',
                         'Casualty_Class',
                         'Sex_of_Casualty',
                         'Age_of_Casualty',
                         'Casualty_Severity',
                         'Casualty_Type',
                         'Number_of_Vehicles',
                         'Number_of_Casualties',
                         'Year',
                         'Month',
                         'Vehicle_Type',
                         'Vehicle_Year',
                         'Sex_of_Driver',
                         'Age_of_Driver'
                        ]]
uk_data_df.head()
```

```python
# Creating dictionaries for all domain values for each column that needs it, 
# Also ensuring that the terminology used here is consistent between both Canada and UK data

casualty_class = {
    1:"Driver",
    2:"Passenger",
    3:"Pedestrian"
}

gender = {
    1:"Male",
    2:"Female",
    3:"Undisclosed",
    -1:"Undisclosed"
}

severity = {
    1:"Fatal",
    2:"Non-Fatal",
    3:"Non-Fatal"
}

casualty_type = {
    0:"Pedestrian",
    1:"Bicyclist",
    2:"Motorcyclist",
    3:"Motorcyclist",
    4:"Motorcyclist",
    5:"Motorcyclist",
    8:"Motor Vehicle Occupant",
    9:"Motor Vehicle Occupant",
    10:"Motor Vehicle Occupant",
    11:"Motor Vehicle Occupant",
    16:"Horse Rider",
    17:"Motor Vehicle Occupant",
    18:"Motor Vehicle Occupant",
    19:"Motor Vehicle Occupant",
    20:"Motor Vehicle Occupant",
    21:"Motor Vehicle Occupant",
    22:"Motorcyclist",
    23:"Motorcyclist",
    90:"Undisclosed",
    97:"Motorcyclist",
    98:"Motor Vehicle Occupant"
}

vehicle = {
    0:"Pedestrian",
    1:"Bicycle",
    2:"Motorcycle",
    3:"Motorcycle",
    4:"Motorcycle",
    5:"Motorcycle",
    8:"Passenger Vehicle",
    9:"Passenger Vehicle",
    10:"Bus",
    11:"Bus",
    16:"Horse",
    17:"Agriculture Machine",
    18:"Tram",
    19:"Cargo Vehicle",
    20:"Cargo Vehicle",
    21:"Cargo Vehicle",
    22:"Motorcycle",
    23:"Motorcycle",
    90:"Undisclosed",
    97:"Motorcycle",
    98:"Undisclosed"    
}

uk_data_df["Casualty_Class"] = uk_data_df["Casualty_Class"].map(casualty_class)
uk_data_df["Sex_of_Casualty"] = uk_data_df["Sex_of_Casualty"].map(gender)
uk_data_df["Casualty_Severity"] = uk_data_df["Casualty_Severity"].map(severity)
uk_data_df["Casualty_Type"] = uk_data_df["Casualty_Type"].map(casualty_type)
uk_data_df["Vehicle_Type"] = uk_data_df["Vehicle_Type"].map(vehicle)
uk_data_df["Sex_of_Driver"] = uk_data_df["Sex_of_Driver"].map(gender)
```

```python
# Further cleaning up the columns needed

uk_data_clean = uk_data_df[['Year',
                             'Month',
                             'Casualty_Type',
                             'Casualty_Severity',
                             'Sex_of_Casualty',
                             'Age_of_Casualty',
                             'Number_of_Vehicles',
                             'Vehicle_Type',
                             'Vehicle_Year',
                             'Sex_of_Driver',
                             'Age_of_Driver',
                             'Accident_Index'
                            ]]
uk_data_clean.head()
```

```python
# Renaming columns so it is more descriptive

uk_data_clean = uk_data_clean.rename(columns = {"Year" : "Collision_Year",
                                                  "Month" : "Collision_Month",
                                                  "Casualty_Severity" : "Casualty_Severity",
                                                  "Sex_of_Driver" : "Driver_Gender",
                                                  "Sex_of_Casualty" : "Casualty_Gender",
                                                  "Casualty_Type" : "Casualty_Position",
                                                  "Age_of_Driver" : "Driver_Age",
                                                  "Age_of_Casualty" : "Casualty_Age"
})
uk_data_clean.head()
```

```python
# Narrowing field of data to encompass data between beginning of 2009, and end of of 2014

uk_data_fin_df = uk_data_clean.loc[(uk_data_clean["Collision_Year"]>2008) & (uk_data_clean["Collision_Year"]<2015),:]
uk_data_fin_df.head()
```

```python
# Converting final dataframe into a new CSV to stage for loading process

uk_data_fin_df.to_csv(os.path.join("..","Staging","UK_Final.csv"), index = False, header=True)
```
