```python
import os
import pandas as pd
```

```python
uk_accident_file = os.path.join("..","source_unzip","UK","Accidents0515.csv")
uk_casualty_file = os.path.join("..","source_unzip","UK","Casualties0515.csv")
uk_vehicle_file = os.path.join("..","source_unzip","UK","Vehicles0515.csv")
```

```python
uk_accident_df = pd.read_csv(uk_accident_file)
uk_accident_df.head()
```

```python
uk_casualty_df = pd.read_csv(uk_casualty_file)
uk_casualty_df.head()
```

```python
uk_vehicle_df = pd.read_csv(uk_vehicle_file)
uk_vehicle_df.head()
```

```python
uk_acci_casu_df = pd.merge(uk_casualty_df,uk_accident_df,on="Accident_Index",how="left")
uk_acci_casu_df.head()
```

```python
uk_acci_casu_vehi_df = pd.merge(uk_acci_casu_df,uk_vehicle_df,on="Accident_Index",how="left")
uk_acci_casu_vehi_df.head()
```
